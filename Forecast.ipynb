```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/Q3_Sikka data.csv")
```

```python
import datetime as dt
data["Date"] = pd.to_datetime(data["Date"])
data['Date']=data['Date'].map(dt.datetime.toordinal)

data.head()
```

```python
data.isnull().sum()
```

```python
test = data[data.Referrals.isnull()]
test.head()
```

```python
train = data.dropna()
train.head()
```

```python
X_train = train.drop(['Referrals'], axis = 1 )
y_train = train.Referrals
X_test = test.drop(['Referrals'], axis=1)
```

```python
X_train.shape, y_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
prediction = lm.predict(X_test)
prediction
```

```python
test["Referrals"] = prediction
test.head()
```
